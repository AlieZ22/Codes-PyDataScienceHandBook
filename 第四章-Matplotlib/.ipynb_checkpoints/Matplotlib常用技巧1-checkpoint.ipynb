### 两种画图接口
这是容易让人混淆的，Matplotlib有两种画图接口
一种是便捷的MATLAB风格接口
另一个是功能更强大的面向对象接口

#### MATLAB风格接口
这种接口的重要特点是 **有状态的(stateful)**:它会持续跟踪“当前的”图形和坐标轴。可以用plt.gcf()（获取当前图形）和plt.gca()（获取当前坐标轴）来查看基本信息。
<br>但如何从第二个图，回到第一个图呢？基于MATLAB风格实现比较复杂，但还有更好的办法：面向对象接口

```python
plt.figure()    # 创建图形
# 创建两个子图的第一个，设置坐标轴
plt.subplot(2,1,1) # 行，列，子图编号
plt.plot(x,np.sin(x))    
# 创建两个子图的第一个，设置坐标轴
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

#### 面向对象接口
适应更复杂的场景，更好地控制自己的图形。
面图函数不再受到当前“活动”图形或坐标轴的限制，而变成了显示的Figure和Axes的方法。

```python
# 先创建图形网格
# ax是一个包含两个Axes对象的数组
fig,ax = plt.subplots(2)

# 在每个对象上调用plot()方法
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

#### 调整图形：线条的颜色与风格
plt.plot()函数可以通过相应参数设置颜色与风格。修改颜色使用color参数。调整线条风格用linestyle参数。

```python
# 控制图形元素的颜色
plt.plot(x,np.sin(x-0),color='blue')    # 标准颜色名称
plt.plot(x,np.sin(x-1),color='g')       # 缩写代码(rgbcmyk)
plt.plot(x,np.sin(x-2),color='0.75')    # 范围在0~1的灰度值
plt.plot(x,np.sin(x-3),color='#FFDD44') # 十六进制
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3)) # RGB元组
plt.plot(x,np.sin(x-5),color='chartreuse')  # HTML颜色名称
# 如果不指定颜色，那么Matplotlib就会为多条线自动循环使用一组默认的颜色
```

```python
# 调整线条的风格
plt.plot(x,x+0,linestyle='solid')   # '-' 实线
plt.plot(x,x+1,linestyle='dashed')  # '--' 虚线
plt.plot(x,x+2,linestyle='dashdot') # '-.' 点划线
plt.plot(x,x+3,linestyle='dotted')  # ':' 实点线
```

```python
# 也可以将linestyle和color编码组合起来使用
plt.plot(x,x+0,'-g')
plt.plot(x,x+1,'--c')
plt.plot(x,x+2,'-.k')
plt.plot(x,x+3,':r')
```

#### 调整图形：坐标轴上下限
虽然Matplotlib会自动选择合适的坐标轴上下限，但有时候自定义更好。调整坐标轴上下限最基础的方法是plt.xlim()和plt.ylim()

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
# 也可以逆序设置：xlim(10,0),ylim(1.5,-1.5)
```

```python
# 或者也可以用plt.axis()方法，传入[xmin,xmax,ymin,ymax]
plt.plot(x,np.cos(x))
plt.axis([-1,11,-1.5,1.5])
plt.axis('tight')   # 自动收紧坐标轴，不留空白
```

#### 设置图形标签
这部分内容包括：图形标题，坐标轴标题，简易图例

```python
plt.plot(x,np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# 在单个坐标轴上显示多条线时，创建图例显示每条线是很有效的方法
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.axis('equal')

plt.legend()
```

```python
# Matplotlib陷阱
# 绝大多数plt函数可以直接转化为ax方法，但并非所有的命令都可以这样用
# 如plt.xlabel()  --> ax.set_xlabel()
# 通常在用面向对象接口画图时，不要单独调用这些函数，而是采用ax.set()方法，一次性设置所有属性
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),xlabel='x',ylabel='sin(x)',title='A Simple Plot')
```

### 简易散点图(scatter plot)
与线型图类似，这种图形不再由线段连接，而是由独立的点、圆圈或其他形状构成。

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# 用plt.plot画散点图
x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'o',color='black')   # 第三个参数表示图形符号的类型
```

```python
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','^','<','>','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),marker,label="marker={}".format(marker))
plt.legend(numpoints=1)
plt.xlim(0,1.8)
```

```python
# 这些代码还可以与线条，颜色代码组合起来，画出一条连接散点的线
plt.plot(x,y,'-ok')    # 直线'-' , 圆圈'o' , 黑色'k'
```

```python
# plt.plot支持许多设置线条和散点属性的参数
plt.plot(x,y,'-p',color='gray',
        markersize=15,linewidth=4,
        markerfacecolor='white',markeredgecolor='gray',
        markeredgewidth=2)
```

```python
# 用plt.scatter画散点图
plt.scatter(x,y,marker='o')
```

**plt.scatter与plt.plot的主要区别在于**，前者在创建散点图时具有更高的灵活性，可以单独控制每个散点与数据匹配，也可以让每个散点具有不同的属性（大小，表面颜色，边框颜色等）

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')   # alpha表示透明度
plt.colorbar() # 显示颜色条
```

#### 实例演示：Scikit-Learn程序库中的鸢尾花数据
有三种鸢尾花，其花瓣（petal)与花萼（sepal)的长度和宽度都经过了仔细测量
<br>散点图可以让我们同时看到不同维度的数据：每个点的坐标值(x,y)分别表示花萼的长度和宽度，而点的大小表示花瓣的宽度，三种颜色对应于三种不同类型的鸢尾花。这类多颜色与多特征的散点图在探索和演示数据时非常有用。

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T    # (4,150) ndarray

plt.scatter(features[0],features[1],alpha=0.2,s=100*features[3],c=iris.target,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

#### plot与scatter的效率对比
在数据量较小的时候，两者在效率上的差异不大。
但是当数据变大到几千个散点时，plt.plot的效率将大大高于plt.scatter。<br>这是因为**plt.scatter会对每个散点进行单独的大小与颜色的渲染**，因此渲染器会消耗更多的资源；而在plt.plot中，散点基本都彼此复制，整个数据集中所有点的颜色、尺寸都只需要配置一次。

### 可视化异常处理
在数据可视化的结果中用图形将误差有效地显示出来，就可以提供充分的信息

#### 基本误差线(errorbar)

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x,y,yerr=dy,fmt='.k')
```

```python
# errorbar还有许多改善结果的选项，可以自定义误差线图形的绘画风格
# 尽量让误差线的颜色比数据点的颜色浅一点
plt.errorbar(x,y,yerr=dy,fmt='o',color='black',ecolor='lightgray',elinewidth=3,capsize=0)
```

#### 连续误差
有时候可能需要显示连续变量的误差。Matplotlib中没有内置的简便方法可以解决，但是用plt.plot与plt.fill_between来解决也不是很难。

```python
from sklearn.gaussian_process import GaussianProcess
# 定义模型和要画的数据
model = lambda x:x*np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

# 计算高斯过程拟合结果
gp = GaussianProcess(corr='cubic',theta0=1e-2,thetaL=1e-4,thetaU=1E-1,random_start=100)
gp.fit(xdata[:,np.newaxis],ydata)
xfit = np.linspace(0,10,1000)
yfit, MSE = gp.predict(xfit[:,np.newaxis],eval_MSE = True)
dyfit = 2* np.sqrt(MSE)  # 2*sigma~95%置信区间

# 将结果可视化
plt.plot(xdata,ydata,'or')
plt.plot(xfit,yfit,'-',color='gray')
plt.fill_between(xfit,yfit-dyfit,yfit+dyfit,color='gray',alpha=0.2)
plt.xlim(0,10)
```

通过区域填充来表示连续误差。<br>fill_between函数设置为：首先传入x轴坐标值，然后传入y轴下边界以及y轴上边界，这样整个区域就被误差线填充了。

### 密度图和等高线图
有时在二维图上用等高线图或者彩色图来表示三维数据是个不错的方法，Matplotlib提供三个函数解决这个问题。<br>
1，plt.contour画等高线图<br>
2，plt.contourf画带有填充色的等高线图的色彩<br>
3，plt.imshow显示图形

#### 三维函数的可视化

```python
# 首先用函数z = f(x,y)演示一个等高线图
def f(x,y):
    return np.sin(x)**10+np.cos(10+y*x)*np.cos(x)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contour(X,Y,Z,color='black')
```

注意：当图形中只使用一种颜色时，默认使用虚线表示负数，使用实线表示正数。<br>
另外，可以用cmap参数设置一个线条配色方案来自定义颜色

```python
plt.contour(X,Y,Z,20,cmap='RdGy')   # 红-灰,Red-Gray缩写
```

虽然这幅图看起来好像漂亮一些，但是线条之间的间隙还是有点大。<br>
我们通过plt.contourf()函数来填充等高线图。

```python
plt.contourf(X,Y,Z,20,cmap='RdGy')
plt.colorbar()
```

```python
# 颜色的改变是一个离散而非连续的过程，我们可以用plt.imshow()函数来处理，渲染成渐变图
plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

#### 等高线图与彩色图的组合
需要一幅背景色半透明的彩色图（可以通过alpha参数设置透明度）<br>
与另一幅坐标轴相同、带数据标签的等高线叠放在一起（用plt.clabel()函数实现）

```python
contours = plt.contour(X,Y,Z,3,colors='black')
plt.clabel(contours,inline=True,fontsize=8)
plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap='RdGy',alpha=0.5)
plt.colorbar()
```

### 频次直方图、数据区间划分和分布密度

```python
# 仅用一行代码就可以创建一个简易的频次直方图
data = np.random.randn(1000)
plt.hist(data)
```

```python
# 调整显示结果
plt.hist(data,bins=30,normed=True,alpha=0.5,
        histtype='stepfilled',color='steelblue',
        edgecolor='none')
```

在用频次直方图对不同分布特征的样本进行对比时，将histtype='stepfilled'与透明性设置参数alpha结合使用的效果很好

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)

kwargs = dict(histtype='stepfilled',alpha=0.3,normed=True,bins=40)
plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs)
```

如果只需要简单地计算频次直方图（计算每段区间的样本数），而不想画图显示，那么可以直接用np.histogram()

```python
counts,bin_edges = np.histogram(data,bins=5)
print(counts)
```

#### 二维频次直方图与数据区间划分
将二维数组按照二维区间进行切分，创建二维频次直方图

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,10000).T

plt.hist2d(x,y,bins=30,cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

**plt.hexbin:六边形区间划分**<br>
二维频次直方图是由坐标轴正交的方块分割而成，还有一种常用的方式是用正六边形分割，将二维数据集分割成蜂窝状。

```python
plt.hexbin(x,y,gridsize=30,cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

#### 核密度估计
评估多维数据分布密度的常用方法是核密度估计（kernel density estimation,KDE)

```python
from scipy.stats import gaussian_kde

# 拟合数组维度[Ndim,Nsamples]
data = np.vstack([x,y])
kde = gaussian_kde(data)

# 用一对规则的网格数据进行拟合
xgrid = np.linspace(-3.5,3.5,40)
ygrid = np.linspace(-6,6,40)
Xgrid,Ygrid = np.meshgrid(xgrid,ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(),Ygrid.ravel()]))

# 画出结果
plt.imshow(Z.reshape(Xgrid.shape),
          origin='lower',aspect='auto',
          extent=[-3.5,3.5,-6,6],
          cmap='Blues')
cb = plt.colorbar(label='density')
```
